```python
import gymnasium as gym
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
env = gym.make("CarRacing-v2", render_mode = "human")
```

```python
env.reset()
```

```python
counter = 1000
capture = 0

while counter != 0:
    env.render()
    env.step(env.action_space.sample())
    capture = env.observation_space
    print(counter)
    counter -= 1
```
